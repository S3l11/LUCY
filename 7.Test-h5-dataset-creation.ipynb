```python
path_test_list = "./test-samples.txt"
path_test_file = open(path_test_list, 'r')
test_cts = path_test_file.readlines()

N_SLICES = 250
RES_Y = 190
RES_X = 270

N_SAMPLES = len(test_cts)

ds_test_fp = np.memmap("./test-samples-memmap.dat", mode='w+', shape=(N_SAMPLES, N_SLICES, RES_Y, RES_X))
ds_test_lab = np.zeros((N_SAMPLES, 2))

i = 0
seed(29)
print("N. of test samples: ",N_SAMPLES)
for path in test_cts:
  i += 1

  path = path.strip()

  print(i)
  #print(path)

  scans = np.load(path)

  ds_test_fp[i-1] = scans

  if "LUAD" in path:
    print("LUAD")
    ds_test_lab[i-1, 0] = 1
  else:
    print("LUSC")
    ds_test_lab[i-1, 1] = 1

  print(ds_test_lab[i-1])

  # Show some slices
  plt.imshow(scans[125, :, :], cmap=plt.cm.gray, vmin=0, vmax=80)
  plt.show()
```

```python
import h5py

ds_test_h5f = h5py.File('./test-zyx-250x190x270.h5', 'w')
```

```python
ds_test_h5f.create_dataset('test_X', data=ds_test_fp)
ds_test_h5f.create_dataset('test_Y', data=ds_test_lab)
```

```python
ds_test_h5f.close()
```

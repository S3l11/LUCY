```
import numpy as np
import os
import scipy.ndimage
import glob
import matplotlib.pyplot as plt
```

```
def sample_stack(stack, rows=6, cols=6, start_with=20, show_every=2):
    fig,ax = plt.subplots(rows,cols,figsize=[12,12])

    for i in range(rows*cols):
        ind = start_with + i*show_every
        ax[int(i/rows),int(i % rows)].set_title('slice %d' % ind)
        ax[int(i/rows),int(i % rows)].imshow(stack[ind],cmap='gray')
        ax[int(i/rows),int(i % rows)].axis('off')

    plt.show()
```

# LUAD (also called ADC)

### NSCLC-Radiogenomics

```
files = []
path_preproc_scans = ".../" # <--- INSERT HERE (...) THE PATH OF PREPROCESSED LUAD SCANS
for r, d, f in os.walk(path_preproc_scans):
  for file in f:
    if file.endswith('.npy'):
      files.append(os.path.join(r, file))
```

```
i = 0
for file in range (len(files)):
  scan = np.load(files[i])
  scan = scan.astype(int)
  scan_reshaped = np.transpose(scan, (2, 1, 0)) # x, y, z
  nonzeros = np.where(scan_reshaped != 0)
  scan_new = scan_reshaped[min(nonzeros[0]) : max(nonzeros[0]) + 1, min(nonzeros[1]) : max(nonzeros[1]) + 1]
  scan_new_zyx = np.transpose(scan_new, (2, 1, 0)) # z, y, x

  cropped_scan = []
  for img in scan_new_zyx:
    if not np.any(np.isnan(img)) and np.all(np.isfinite(img)): # if there are no NaN or infinite values
      cropped_scan.append(img)

  cut_cropped_scan = []
  for slice in cropped_scan:
    countzero = not np.any(slice)
    if not countzero:
      cut_cropped_scan.append(slice) # append only if it is not totally black
  cut_cropped_scan = np.array(cut_cropped_scan)

  #print("Scan n. "+str(i+1))
  #print("Shape before automated crop and cut\t"+str(scan.shape)+"") # z, y, x
  #print("Shape after automated crop and cut\t"+str(cut_cropped_scan.shape)+"") # z, y, x
  i+=1

  np.save(".../crop-cut-scan-a-"+str(i),cut_cropped_scan) # <--- INSERT HERE (...) THE PATH WHERE TO SAVE CROP&CUT LUAD SCANS
```

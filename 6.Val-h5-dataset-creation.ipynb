```python
path_val_list = "./val-samples.txt"
path_val_file = open(path_val_list, 'r')
val_cts = path_val_file.readlines()

N_SLICES = 250
RES_Y = 190
RES_X = 270

N_SAMPLES = len(val_cts)

ds_val_fp = np.memmap("./val-samples-memmap.dat", mode='w+', shape=(N_SAMPLES, N_SLICES, RES_Y, RES_X))
ds_val_lab = np.zeros((N_SAMPLES, 2))

i = 0
seed(29)
print("N. of val samples: ",N_SAMPLES)
for path in val_cts:
  i += 1

  path = path.strip()

  print(i)
  #print(path)

  scans = np.load(path)

  ds_val_fp[i-1] = scans

  if "LUAD" in path:
    print("LUAD")
    ds_val_lab[i-1, 0] = 1
  else:
    print("LUSC")
    ds_val_lab[i-1, 1] = 1

  print(ds_val_lab[i-1])

  # Show some slices
  plt.imshow(scans[125, :, :], cmap=plt.cm.gray, vmin=0, vmax=80)
  plt.show()
```

```python
import h5py

ds_val_h5f = h5py.File('./val-zyx-250x190x270.h5', 'w')
```

```python
ds_val_h5f.create_dataset('val_X', data=ds_val_fp)
ds_val_h5f.create_dataset('val_Y', data=ds_val_lab)
```

```python
ds_val_h5f.close()
```
